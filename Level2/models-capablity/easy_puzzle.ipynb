```python
import os
import requests
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
load_dotenv(override=True)
google_api_key = os.getenv('GOOGLE_API_KEY')
```

```python
if google_api_key:
    print(f"Google API Key exists and begins {google_api_key[:2]}")
else:
    print("Google API Key not set (and this is optional)")
```

```python
gemini_url = "https://generativelanguage.googleapis.com/v1beta/openai/"
gemini = OpenAI(api_key=google_api_key, base_url=gemini_url)
ollama = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
```

```python
easy_puzzle = [
    {"role": "user", "content":
        "You toss 2 coins. One of them is heads. What's the probability the other is tails? Answer with the probability only."},
]
```

```python
response = gemini.chat.completions.create(model="gemini-2.5-flash", messages=easy_puzzle, reasoning_effort="minimal")
display(Markdown(response.choices[0].message.content))
```

```python
response = gemini.chat.completions.create(model="gemma-4-31b-it", messages=easy_puzzle)
display(Markdown(response.choices[0].message.content))
```

```python
response = ollama.chat.completions.create(model="llama3.2:1b", messages=easy_puzzle)
display(Markdown(response.choices[0].message.content))
```
