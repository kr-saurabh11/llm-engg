```python
import os
import requests
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
load_dotenv(override=True)
google_api_key = os.getenv('GOOGLE_API_KEY')
```

```python
if google_api_key:
    print(f"Google API Key exists and begins {google_api_key[:2]}")
else:
    print("Google API Key not set (and this is optional)")
```

```python
gemini_url = "https://generativelanguage.googleapis.com/v1beta/openai/"
gemini = OpenAI(api_key=google_api_key, base_url=gemini_url)
ollama = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
```

```python
hard = """
On a bookshelf, two volumes of Pushkin stand side by side: the first and the second.
The pages of each volume together have a thickness of 2 cm, and each cover is 2 mm thick.
A worm gnawed (perpendicular to the pages) from the first page of the first volume to the last page of the second volume.
What distance did it gnaw through?
"""
hard_puzzle = [
    {"role": "user", "content": hard}
]
```

```python
response = gemini.chat.completions.create(model="gemini-2.5-flash", messages=hard_puzzle, reasoning_effort="minimal")
display(Markdown(response.choices[0].message.content))
```

```python
response = gemini.chat.completions.create(model="gemma-4-31b-it", messages=hard_puzzle)
display(Markdown(response.choices[0].message.content))
```

```python
response = ollama.chat.completions.create(model="llama3.2:1b", messages=hard_puzzle)
display(Markdown(response.choices[0].message.content))
```
