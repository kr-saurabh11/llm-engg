```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
load_dotenv(override=True)
google_api_key = os.getenv('GOOGLE_API_KEY')
deepseek_api_key = os.getenv('DEEPSEEK_API_KEY')
```

```python
if google_api_key:
    print(f"Google API Key exists and begins {google_api_key[:2]}")
else:
    print("Google API Key not set (and this is optional)")

if deepseek_api_key:
    print(f"DeepSeek API Key exists and begins {deepseek_api_key[:2]}")
else:
    print("DeepSeek API Key not set (and this is optional)")
```

```python
gemini_url = "https://generativelanguage.googleapis.com/v1beta/openai/"
gemini = OpenAI(api_key=google_api_key, base_url=gemini_url)
deepseek_url = "https://api.deepseek.com"
deepseek = OpenAI(api_key=deepseek_api_key, base_url=deepseek_url)
ollama = OpenAI(base_url='http://localhost:11434/v1', api_key='ollama')
```

```python
dilemma_prompt = """
You and a partner are contestants on a game show. You're each taken to separate rooms and given a choice:
Cooperate: Choose "Share" — if both of you choose this, you each win $1,000.
Defect: Choose "Steal" — if one steals and the other shares, the stealer gets $2,000 and the sharer gets nothing.
If both steal, you both get nothing.
Do you choose to Steal or Share? Pick one.
"""

dilemma = [
    {"role": "user", "content": dilemma_prompt},
]
```

```python
response = gemini.chat.completions.create(model="gemini-2.5-flash", messages=dilemma, reasoning_effort="high")
display(Markdown(response.choices[0].message.content))
```

```python
response = gemini.chat.completions.create(model="gemma-4-31b-it", messages=dilemma)
display(Markdown(response.choices[0].message.content))
```

```python
response = ollama.chat.completions.create(model="llama3.2:1b", messages=dilemma)
display(Markdown(response.choices[0].message.content))
```

```python
response = deepseek.chat.completions.create(model="deepseek-v4-flash", messages=dilemma)
display(Markdown(response.choices[0].message.content))
```
