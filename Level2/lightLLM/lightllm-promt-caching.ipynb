```python
from litellm import completion
from IPython.display import display, Markdown
```

```python
with open("hamlet.txt", "r", encoding="utf-8") as f:
    hamlet = f.read()

loc = hamlet.find("Speak, man")
print(hamlet[loc:loc+100])
```

```python
question = [{"role": "user", "content": "In Hamlet, when Laertes asks 'Where is my father?' what is the reply?"}]
```

```python
response = completion(model="gemini/gemma-4-31b-it", messages=question)
display(Markdown(response.choices[0].message.content))
```

```python
print(f"Input tokens: {response.usage.prompt_tokens}")
print(f"Output tokens: {response.usage.completion_tokens}")
print(f"Total tokens: {response.usage.total_tokens}")

# Safe cost extraction with fallback
try:
    response_cost = response._hidden_params.get("response_cost") if hasattr(response, "_hidden_params") else None
    if response_cost is not None:
        print(f"Total cost: {response_cost*100:.4f} cents")
    else:
        print("Total cost: Not available for this model")
except Exception as e:
    print(f"Cost information: Not available ({type(e).__name__})")
```

```python
question[0]["content"] += "\n\nFor context, here is the entire text of Hamlet:\n\n"+hamlet
response = completion(model="gemini/gemma-4-31b-it", messages=question)
display(Markdown(response.choices[0].message.content))
```

```python
print(f"Input tokens: {response.usage.prompt_tokens}")
print(f"Output tokens: {response.usage.completion_tokens}")
print(f"Total tokens: {response.usage.total_tokens}")
print(f"Cached tokens: {response.usage.prompt_tokens_details.cached_tokens}")
```
