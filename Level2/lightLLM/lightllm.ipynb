```python
from litellm import completion
from IPython.display import display, Markdown
```

```python
tell_a_joke = [
    {"role": "user", "content": "Tell a joke for a student on the journey to becoming an expert in LLM Engineering"},
]
```

```python
response = completion(model="gemini/gemma-4-31b-it", messages=tell_a_joke)
reply = response.choices[0].message.content
display(Markdown(reply))
```

```python
print(f"Input tokens: {response.usage.prompt_tokens}")
print(f"Output tokens: {response.usage.completion_tokens}")
print(f"Total tokens: {response.usage.total_tokens}")

# Safe cost extraction with fallback
try:
    response_cost = response._hidden_params.get("response_cost") if hasattr(response, "_hidden_params") else None
    if response_cost is not None:
        print(f"Total cost: {response_cost*100:.4f} cents")
    else:
        print("Total cost: Not available for this model")
except Exception as e:
    print(f"Cost information: Not available ({type(e).__name__})")
```
